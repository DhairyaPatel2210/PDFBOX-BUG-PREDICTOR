### Creating the vector assembler from the dataframe and splitting in the train and test

```
# this block prepares the trainig and testing data from the original datasets
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.evaluation import MulticlassClassificationEvaluator

feature_columns = ['WMC', 'LCOM', 'DIT', 
    'CBO', 'RFC', 'LCAM', 'LOC']
label_column = 'IsBuggy'

# Assemble features
assembler = VectorAssembler(inputCols=feature_columns, outputCol="Features")
filteredDf = assembler.transform(df)

assembler = VectorAssembler(inputCols=feature_columns, outputCol="Features")
test_dataframe = assembler.transform(test_df)

# Splitting the data into train and test
train_data, test_data = filteredDf.randomSplit([0.9,0.1], seed=123)
```

### Decision Tree Classifier and computing the accuracy 

##### Computation by splitting the data into train and test

```
# checking in-built decision tree algorithm to get the idea about the accuracy
from pyspark.ml.classification import DecisionTreeClassifier

dt = DecisionTreeClassifier(labelCol="IsBuggy", featuresCol="Features")
dt_model = dt.fit(filteredDf)

# Make predictions on test data
predictions_dt = dt_model.transform(test_data)
test_predictions = dt_model.transform(test_dataframe)

evaluator = MulticlassClassificationEvaluator(labelCol="IsBuggy", predictionCol="prediction", metricName="accuracy")
accuracy1 = evaluator.evaluate(test_predictions)
accuracy2 = evaluator.evaluate(predictions_dt)

print("Test Accuracy With V2.0.31 Data = {:.2f}%".format(accuracy1 * 100))
print("Test Accuracy With Training Data Collected = {:.2f}%".format(accuracy2 * 100))
```

##### Computation by providing the test data for version 2.0.31 and calculation of Precision and Recall

```
from pyspark.mllib.evaluation import MulticlassMetrics

labeled_df = test_predictions.withColumnRenamed('IsBuggy', 'label')
predictionAndLabels = labeled_df.select("prediction", "label").rdd.map(lambda row: (row.prediction, float(row.label)))

metrics = MulticlassMetrics(predictionAndLabels)

# Overall statistics
precision1 = metrics.precision(label=1)
recall1 = metrics.recall(label=1)
precision0 = metrics.precision(label=0)
recall0 = metrics.recall(label=0)
print("Summary Stats")
print("Precision for label 1 = {:.2f}%".format(precision1 * 100))
print("Recall for label 1 = {:.2f}%".format(recall1 * 100))
print("Precision for label 0 = {:.2f}%".format(precision0 * 100))
print("Recall for label 0 = {:.2f}%".format(recall0 * 100))
```
